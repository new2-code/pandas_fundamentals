Pandas Data Frame:

Pandas DataFrame is two-dimensional tabular data structure with labeled axes (rows and columns).


```python
import pandas as pd
```

Creating pandas Dataframes

```python
from sklearn.datasets import fetch_california_housing
```

```python
cal_house = fetch_california_housing()
```

```python
print(cal_house)
```

```python
print(type(cal_house))
```

```python
#Pandas Dataframe

cal_house_df = pd.DataFrame(data=cal_house.data, columns=cal_house.feature_names)
```

```python
cal_house_df.head()
```

```python
cal_house_df.shape
```
